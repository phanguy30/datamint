```python
from Value import Value
import MLPnetwork

model = MLPnetwork.MLPNetwork(2, [4, 4, 1], activation_type="relu")

for step in range(1000):
    model.zero_grad()

    out = model.forward([Value(1), Value(-1)])   # out is a list of Value
    y_pred = out[0]
    y_true = Value(1)

    loss = (y_pred - y_true)**2
    loss.backward()

    for p in model.parameters():
        p.data -= 0.01 * p.grad
```
